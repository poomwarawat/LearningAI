```python
gray_image = cv2.imread(('pic/coin.jpg'), 0)
```

```python
plt.imshow(gray_image, cmap="gray")
```

```python
ret, thresh = cv2.threshold(gray_image, 0, 255, cv2.THRESH_OTSU)
```

```python
plt.imshow(thresh, cmap="gray")
plt.title("Thresh Image")
plt.axis("off")
```

```python
img1 = thresh.copy()
```

```python
se = cv2.getStructuringElement(cv2.MORPH_RECT, (5, 5))
```

```python
erosion = cv2.erode(img1, se, iterations=1)
```

```python
plt.imshow(erosion, cmap="gray")
plt.title("Erosion Image")
plt.axis("off")
```

```python
dilation = cv2.dilate(img1, se, iterations=1)
```

```python
plt.imshow(dilation, cmap="gray")
plt.title("Dilation Image")
plt.axis("off")
```

```python
opening = cv2.morphologyEx(img1, cv2.MORPH_CLOSE, se)
```

```python
plt.imshow(opening, cmap="gray")
plt.title("Open Image")
plt.axis("off")
```

```python
closing = cv2.morphologyEx(img1, cv2.MORPH_CLOSE, se)
```

```python
plt.imshow(closing, cmap="gray")
plt.title("closing Image")
plt.axis("off")
```

```python
ret,thresh = cv2.threshold(closing, 127, 255, 0)
contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_NONE)

cv2.drawContours(gray_image, contours, -1, (255,0,0), 3)
plt.imshow(gray_image)
plt.title("Line test")
```
